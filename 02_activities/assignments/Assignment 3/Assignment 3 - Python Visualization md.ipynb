    > What software did you use to create your data visualization?
    Python was used to create the visualization for Revenue/Hotel Room and Occupancy Rates in the GTA for Dec 2024.

    > Who is your intended audience?
    The intended audience for this visualization would be anyone interested in the average revenue of hotel rooms in the GTA during one of the holiday periods. This may include hotel owners, investors, and government officials within the GTA. 
    
    > What information or message are you trying to convey with your visualization? 
    The information that this visualization is trying to convey is comparing the different average revenue amounts of hotel rooms for various locations in the GTA. Seeing if there are any differences in revenue based on location. Additionally, we are also looking at the average occupancy rates for the hotels in each region within the GTA to provide some context on the hotel's foot traffic.  

    > What aspects of design did you consider when making your visualization? How did you apply them? With what elements of your plots? 
    I considered: 
        - minimizing cognitive load: trying to maintain a clean layout by using simple colours and simple shapes for the bar chart and avoid overplotting.
        - layering and comparison by showing two similar metrics (revenue and occupany) in different visuals but on the same plot to allow for easy comparisons.

    > How did you ensure that your data visualizations are reproducible? If the tool you used to make your data visualization is not reproducible, how will this impact your data visualization? 
    To ensure reproducibility in python, we set a random seed so that the random generator is similar for each trial. However, due to the dataset chosen, importing the data directyl was not possible so arrays were made and filled manually for each variable. 

    > How did you ensure that your data visualization is accessible?  
    Redundant encoding such as clear labels with legends were used. Colours that were hihg contrast were used to depict differences. Also keeping the design simple and clear.

    > Who are the individuals and communities who might be impacted by your visualization?  
    Individuals who are involved with the hotel/hospitality industry within the GTA may be impacted. Additionally, individuals who plan to stay at hotels within the GTA may also be affected by this visualization, depending on if any changes are prompted by this visualization.

    Based on what was available from the dataset, I chose to incorporate similar features (revenue and room occupancy) while exluding features that did not relate to each other (interested in the GTA vs other parts of Ontario)
