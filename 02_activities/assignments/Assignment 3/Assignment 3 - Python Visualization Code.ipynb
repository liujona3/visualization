```python
# Import/load libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy
import PIL
import requests
```

```python
np.random.seed(60) #set random seed for reproducibility
x = np.array(["Downtown Toronto", "Toronto Airport", "West", "East/North"]) #manually setting x variable, GTA locations
y1 = np.array([171.94, 113.28, 91.27, 95.62]) #manually setting first y variable, revenue per room
y2 = np.array([0.615, 0.646, 0.602, 0.613]) #manually setting second y variable, room occupancy rates
```

```python
fig, ax1 = plt.subplots(figsize=(10, 5))

ax1.bar(x, y1, color = "grey")
#setting the title and axes (artists)
ax1.set_title('Hotel Revenue per Room in the GTA (Dec 2024)')
ax1.set_xlabel('GTA Locations')
ax1.set_ylabel('Average Revenue per Room ($)')

#create second y-axis
ax2 = ax1.twinx()
ax2.plot(x, y2, 
         color="blue", #make the occupancy plot blue 
         marker="o", #marker at each x will be a circle
         label="Occupancy Rate") #create legend for occupancy rate
ax2.set_ylabel("Occupancy Rate (%)")
ax2.set_ylim(0, 100) #making the range of the second axis 0-100 to represent %
ax2.legend(loc='upper right') #move the legend to the top right

plt.tight_layout()
plt.show()
```
