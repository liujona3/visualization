Excel Visualization - https://data.ontario.ca/dataset/hotel-statistics/resource/08b25430-3326-4317-aa5e-84f96d9935b0 
    > What software did you use to create your data visualization?
    Excel was used to create the visualization for Hotel Rates in the GTA for Dec 2024 and 2023.

    > Who is your intended audience?
    The intended audience of this visualization are people who are interested in the average daily rates of hotels in the GTA during the Dec. Specifically people who are planning to visit the GTA during one of the holiday periods in Dec.
    
    > What information or message are you trying to convey with your visualization? 
    The visualization is trying to convey the price differences of hotels at various locations in the GTA. Additionally the visualization will compare the price in Dec for 2024 and 2023 to provide additionaly context pricing range. 

    > What aspects of design did you consider when making your visualization? How did you apply them? With what elements of your plots?
    I considered: 
        - minimizing cognitive load: trying to maintain a clean layout by using simple colours and simple shapes for the bar chart and avoid overplotting.
        - layering and comparison by showing two similar metrics (years) in different colours to allow for easy comparisons.
    
    > How did you ensure that your data visualizations are reproducible? If the tool you used to make your data visualization is not reproducible, how will this impact your data visualization? 
    Excel was used, so it was not reproducible. This would not impact the data visualization since there are no elements where random numbers need to be generated.

    > How did you ensure that your data visualization is accessible?  
    Redundant encoding such as clear labels with legends were used. Colours that were hihg contrast were used to depict differences. Also keeping the design simple and clear.

    > Who are the individuals and communities who might be impacted by your visualization?  
    Individuals who might be impacted by this visualization would be people who plan to visit the GTA in Dec as they would be provided the avrage rates for hotels during that time. Addiitonally, those involved in hospitality such as AirBnB within the GTA might be impacted.  

    > How did you choose which features of your chosen dataset to include or exclude from your visualization? 
    Based on what was available from the dataset, I chose to incorporate similar features (Average Daily rates, year, GTA) while exluding features that did not relate to each other (interested in the GTA vs other parts of Ontario, revenue)
