```python
# load libraries

import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
# load data, then plot with seaborn package

tips = sns.load_dataset("tips") #dataset within seaborn
print(tips)
```

```python
sns.set_style('whitegrid') #add white grids
#sns.set_style('darkgrid') #can use dark grids

sns.lineplot(data=tips, x='total_bill', y='tip')
```

```python

fig = plt.subplots(figsize = (10,3))

tipgraph = sns.lineplot(data=tips, x='total_bill', y='tip',
                        color = 'hotpink',
                        linestyle = '--',
                        linewidth = 3,
                        marker = 'o',
                        markerfacecolor = 'indigo'
                        )
```

```python
tipgraph = sns.lineplot(data=tips, x='total_bill', y='tip',
                        style = 'time',
                        hue = 'day',
                        palette = ['purple', 'hotpink', 'deepskyblue', 'yellowgreen']
                        )

tipgraph.set(title='Tips vs. Total Bill', 
             xlabel='Total Bill ($)', 
             ylabel='Tip Amount ($)')
```

```python
sns.pairplot( data = tips, hue = 'day')
```

```python
daysplot = sns.relplot( data=tips, x="total_bill", y="tip", 
                       hue="sex", #defines the colour of the data points based on sex variable
                       col="day", #creates unique plots based on the values of the day column
                       kind="scatter", #specify scatter plot
                       col_wrap=2) #specify 2 plots per row
```

```python
# PLOTLY

import plotly.graph_objects as go # 'go' is 'graph objects’
from wordcloud import WordCloud
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted
```

```python
x1 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
graph = go.Figure() #set figure; graph object into variable graph
graph.add_trace(go.Bar(x=x1, y=y1)) #set it as bar graph using addtrace
```

```python
graph.update_layout(title="Pirate Scores",  #updating the graph with titles
                    xaxis_title="Pirates", 
                    yaxis_title="Score") #graph can be interacted, dynamic form as html file; path needs to be chagned from the code in the slide to be saved properly
```

```python
graph = go.Figure()
graph.add_trace(go.Scatter(x=x1, y=y1,
                           mode='markers', #for scatterplot
                           marker=dict( #changing visual asthetics of the graph
                               size=15, #size of dots/points
                               color='hotpink',
                               opacity=1, #point transparency
                               line=dict(width=5, color='purple') #the point outline
                           )))

graph.update_layout( 
    title='Interactive Pirate Plot', 
    xaxis_title='Pirates', 
    yaxis_title='Scores', 
    width=500, height=500)
```

```python
#wordclouds

from wordcloud import WordCloud
df = pd.read_csv("https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv",
                 on_bad_lines='skip')
df
```

```python
# join all our text from each row from quote column into string
text = " ".join(each for each in df.quote)

# generate our wordcloud image
wordcloud = WordCloud(background_color="white",
                      colormap = 'inferno').generate(text)

# use matplotlib syntax to put our image in a figure
figurefig, ax = plt.subplots(figsize=(7, 3))
ax.imshow(wordcloud, #imshow also used to add picture from web onto figure from before
          interpolation='bilinear') #smooths image
ax.axis("off")
```

```python
#venn diagrams
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted

#define sets
A = set(["apple", "banana", "watermelon"])
B = set(["pumpkin", "blueberry", "apple", "key lime"])
```

```python
diagram = venn2_unweighted([A, B], 
                           set_labels = ('Fruits', 'Pies'), 
                           set_colors=("blue", "red"), 
                           alpha=0.5)

#without this, we only get counts. This code gives us our labels inside the venn diagram
diagram.get_label_by_id("10") .set_text("\n".join(A - B)) #\n to get them on separate lines
diagram.get_label_by_id("11") .set_text("\n".join(A & B))
diagram.get_label_by_id("01") .set_text("\n".join(B - A))
```
