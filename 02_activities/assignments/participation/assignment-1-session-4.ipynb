```python
# Import/load libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy
import PIL
import requests
```

```python
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75,50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110, 180, 240, 99, 220])
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, 
                               nrows=1, 
                               figsize=(7, 3))
ax1.scatter(x1,y1, color = 'red')
ax2.bar(x2,y2, color = 'pink')
fig.show()
```

```python
#subplots
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'], 
                                    ['ax2', 'ax3']], 
                                    figsize=(7, 4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)

#setting labels for each plot
someaxes["ax1"].set_xlabel('A Big Label',fontsize=18)
someaxes["ax2"].set_xlabel('Another Label',fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize=18 )
plt.show()
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'], 
                                    ['ax2', 'ax3']], 
                                    figsize=(7, 4),
                                    layout = "constrained") #use layouts to ensure no cutoffs or uneven squishing
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)

#setting labels for each plot
someaxes["ax1"].set_xlabel('A Big Label',fontsize=18)
someaxes["ax2"].set_xlabel('Another Label',fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize=18 )
plt.show()
```

```python
# sample data
x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])

# define one figure and axis
fig, ax = plt.subplots(figsize=(7, 3))

# now call both bar and lineplot on to the same axes
ax.bar(x, y1,color = "indigo")
ax.plot(x, y2,color = "red")
```

```python
#caluclate sd
y2_sd = np.std(y2)

#define lineplot
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")
ax.errorbar(x, #x values 
            y2, #y values 
            yerr = y2_sd, 
            fmt = "none", #format, none = no datapoints, just want r
            ecolor= "indigo", #changing appearance of the error bar
            elinewidth= 4, 
            capsize = 6, 
            capthick= 4,
            errorevery = 2) #error bars every alternate value
```

```python
#adding images to plots
#using images from internet, need ot load additional libraries
from PIL import Image # to open images
import requests # to get images from URLs
from io import BytesIO # to store images
```

```python
#using images from internet
response = requests.get('https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png')
image_file = BytesIO(response.content)
image = Image.open(image_file)
```

```python
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")

#overlay a new axes ontop of the lineplot axes
ax_image = fig.add_axes([0.1, # x coordinate (ON FIGURE, NOT AXES) 
                         0.11, # y coordinate (ON FIGURE, NOT AXES) 
                         0.15, # image width of the original image
                         0.35]) # image height
ax_image.imshow(image) #shows image
ax_image.axis('off') #remove border
```
